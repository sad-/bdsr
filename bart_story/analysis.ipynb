```python
def station_get_value(station, **args):
    if "prop" in args:
        prop = args["prop"]
    else:
        prop = "name"
    values = stations[stations.abbr == station][prop].values
    return values[0]
```

```python
def make_scaler(column):
    mymin = min(column)
    mymax = max(column)
    def retfunc(x):
        return (x - mymin) / (mymax-mymin+1e-8)
    return retfunc
texts = []

def draw_stations(m, station, scaler, annotate=False):
    x,y = m(station.longitude, station.latitude)
    m.plot(x, y, 'o', markersize=0.001*incomes[station.abbr], color='#444444', alpha=0.1+0.9*scaler(incomes[station.abbr]))
    if annotate:
        texts.append(plt.text(x,y,station.abbr))
```

## Part 4 - Distance vs Fare

```python
def crow_distance(station1, station2):
    station1_lat = station_get_value(station1, prop='latitude')
    station1_long = station_get_value(station1, prop='longitude')
    station2_lat = station_get_value(station2, prop='latitude')
    station2_long = station_get_value(station2, prop='longitude')
    return vincenty(geopy.Point(station1_lat, station1_long),
                        geopy.Point(station2_lat, station2_long)).miles
```

```python
station_mean_distance = {}
for station in stations_list:
    dists = []
    for other_station in stations_list:
        if other_station != station:
            dists.append(crow_distance(station, other_station))
    station_mean_distance[station] = np.mean(dists)
```

```python
def find_distance(station1, station2):
    if station1 in line1:
        if station2 in line1:
            idx1, idx2 = line1.index(station1), line1.index(station1)
            start, end = min(idx1, idx2), max(idx1, idx2)
            dist = 0
            for i in xrange(start, end):
                dist += crow_distance(line1[i], line1[i+1])
        elif station2 in line2:
            start1, end1 = line1.index(station1), len(line1) - 1
            start2, end2 = line2.index(station2), len(line2) - 1
            dist = find_distance(line1[start1], line1[end1]) + find_distance(line2[start2], line2[end2])
        elif station2 in line3:
            start1, end1 = line1.index(station1), len(line1) - 1
            start2, end2 = 0, line3.index(station2)
            dist = find_distance(line1[start1], line1[end1]) + find_distance(line3[start2], line3[end2])
        elif station2 in line4:
            start1, end1 = line1.index(station1), len(line3) - 1
            start2, end2 = line4.index(station2), len(line4) - 1
            dist = find_distance(line1[start1], line3[end1]) + find_distance(line4[start2], line4[end2])
        elif station2 in line5:
            start1, end1 = line1.index(station1), len(line3)
            dist = find_distance(line1[start1], line1[end1]) + find_distance(line3[start2], line3[end2]) + find_distance(line4[start3], line4[end3])
        
```

```python
def project_population_for_zipcode(zipcode):
    matching_rows = demographics[demographics.zipcode==zipcode]
    population = list(matching_rows.totalpop)
    avg_increase = sum([population[i] - population[i-1] for i in xrange(1, len(population))])/3
    return population[-1] + avg_increase
```

```python
## aggregate with demographics data
years = [2011, 2012, 2013, 2014, 2015]
populations = {}
for station in station_numbers:
    station_pop = []
    for year in years:
        if year == 2015:
            population = project_population_for_zipcode(station_get_value(station, prop="zipcode"))
        else:
            population = demographics[demographics.zipcode==station_get_value(station, prop="zipcode")]
            population = population[population.year == year].totalpop
        station_pop.append(population)
    populations[station] = station_pop
    #station_numbers[station][j] = station_numbers[station][j]/(30*population)
```

```python
# learn pricing model
distance_array = np.array([augmented_fares.fare, augmented_fares.dist]).T
```

```python
d = 2
x= tf.placeholder("float", shape = [None, d])
w = tf.Variable(tf.random_normal((d,1), stddev=0.1), name="w")
b = tf.Variable(tf.random_normal((1,1), stddev=0.1), name="b")
y_ = tf.placeholder("float")
y = tf.matmul(x, w) + b
loss = tf.reduce_mean((y - y_)**2)
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
session = tf.Session()
session.run(tf.initialize_all_variables())
with session:
    for i in range(10000):
        if i%1000==0:
            print y.eval(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)}).shape
            print np.array(augmented_fares.entry_rate).shape
            training_loss = loss.eval(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)})
            print "training loss after %d epochs: %g"%(i, training_loss)
        train_step.run(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)})
```

```python
trainer = np.hstack((distance_array,np.ones((distance_array.shape[0],1)) ))
price_model = sklearn.linear_model.LinearRegression()
price_model.fit(trainer,np.array(augmented_fares.entry_rate))
```
