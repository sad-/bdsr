## Bart Stations Trends in General

```python
## get all bart_station names into a dict of abbr to station name
stations_dict = {}
for row in stations.itertuples():
    stations_dict[row[2]] = row[1]
```

```python
def station_get_value(station, **args):
    if "prop" in args:
        prop = args["prop"]
    else:
        prop = "name"
    values = stations[stations.abbr == station][prop].values
    return values[0]
```

```python
## index to station
station_index = {}
index_station = {}
for i, station in enumerate(stations_dict):
    station_index[station] = i
    index_station[i] = station
```

```python
bart_station_incomes = {}
for station in stations_dict.keys():
    station_income = []
    for year in years:
        income = demographics[demographics.zipcode == station_get_value(station, prop='zipcode')]
        income = income[income.year == year].median_household_income.values[0]
        station_income.append(income)
    bart_station_incomes[station] = station_income
```

```python
station_numbers = {}
for station in stations_dict:
    enter_at_station = ridership[ridership.enter_abbr==station]
    station_by_hour = enter_at_station.set_index('datetime')
    station_monthly_entry = station_by_hour['trip_count'].groupby(pd.TimeGrouper('D')).sum().resample("1M")
    station_numbers[station] = station_monthly_entry
```

```python
def project_population_for_zipcode(zipcode):
    matching_rows = demographics[demographics.zipcode==zipcode]
    population = list(matching_rows.totalpop)
    avg_increase = sum([population[i] - population[i-1] for i in xrange(1, len(population))])/3
    return population[-1] + avg_increase
```

```python
## aggregate with demographics data
years = [2011, 2012, 2013, 2014, 2015]
populations = {}
for station in station_numbers:
    station_pop = []
    for year in years:
        if year == 2015:
            population = project_population_for_zipcode(station_get_value(station, prop="zipcode"))
        else:
            population = demographics[demographics.zipcode==station_get_value(station, prop="zipcode")]
            population = population[population.year == year].totalpop
        station_pop.append(population)
    populations[station] = station_pop
    #station_numbers[station][j] = station_numbers[station][j]/(30*population)
```

```python
prices_from = {}
prices = {}
for station in stations_dict:
    price_from_station = []
    fares_to = fares[fares.station1==station]
    for end_station in stations_dict:
        if station != end_station:
            fare = fares_to[fares_to.station2 == end_station].fare
            if len(fare) == 0:
                fares_from = fares[fares.station1==end_station]
                fare = fares_from[fares_from.station2==station].fare
            price_from_station.append((station_index[end_station], fare.values[0]))
            prices[(station, end_station)] = fare.values[0]
    prices_from[station] = sorted(price_from_station, key=lambda x:x[1])
```

```python
# learn pricing model
distance_array = np.array([augmented_fares.fare, augmented_fares.dist]).T
```

```python
d = 2
x= tf.placeholder("float", shape = [None, d])
w = tf.Variable(tf.random_normal((d,1), stddev=0.1), name="w")
b = tf.Variable(tf.random_normal((1,1), stddev=0.1), name="b")
y_ = tf.placeholder("float")
y = tf.matmul(x, w) + b
loss = tf.reduce_mean((y - y_)**2)
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
session = tf.Session()
session.run(tf.initialize_all_variables())
with session:
    for i in range(10000):
        if i%1000==0:
            print y.eval(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)}).shape
            print np.array(augmented_fares.entry_rate).shape
            training_loss = loss.eval(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)})
            print "training loss after %d epochs: %g"%(i, training_loss)
        train_step.run(feed_dict={x:distance_array, y_:np.array(augmented_fares.entry_rate)})
```

```python
trainer = np.hstack((distance_array,np.ones((distance_array.shape[0],1)) ))
price_model = sklearn.linear_model.LinearRegression()
price_model.fit(trainer,np.array(augmented_fares.entry_rate))
```
